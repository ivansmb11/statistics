# Regresión lineal

## Mínimos cuadrados

```python
import numpy as np
from numpy.linalg import inv 
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x = [10, 12, 8, 17, 10, 15, 10, 14, 19, 10, 11, 13, 16, 10, 12]
y = [15, 17, 13, 23, 16, 21, 14, 20, 24, 17, 16, 18, 23, 15, 16]
```

```python
plt.scatter(x, y); plt.xlabel('Gastos'); plt.ylabel('Número de pasajeros');
```

```python
X = np.array([np.ones(len(x)), x]).T
a = inv(X.T @ X) @ X.T @ y    ### Fórmula para minimizar los cuadrados
```

```python
x_predict = np.linspace(0, 29, num=100)
y_predict = a[0] + a[1] * x_predict ### Recta 
```

```python
a[1]
```

```python
plt.scatter(x, y)
plt.xlabel('x'); plt.ylabel('Suscriptores'); plt.plot(x_predict, y_predict, 'c');
```
